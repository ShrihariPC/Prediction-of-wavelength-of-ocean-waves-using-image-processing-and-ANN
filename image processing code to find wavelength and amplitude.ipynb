```python
import cv2
import numpy as np
import os
import re
from textwrap import wrap
from operator import itemgetter
import ast
from csv import writer
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

for oneiter in range(2700):
    try:
        image = cv2.imread("Batch 3/validation/frame"+str(oneiter+10814)+".png")
        #print(image)
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        cv2.imwrite("frame0(gray).png", gray_image)
        inverted_image = 140 - gray_image
        cv2.imwrite("frame0(inverted).png", inverted_image)
        blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
        cv2.imwrite("frame0(blured).png", blurred)
        inverted_blurred = 110 - blurred
        pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
        cv2.imwrite("frame0final.png", pencil_sketch)




        def out_fun():
            image = cv2.imread('frame0final.png')
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            # Set threshold level
            threshold_level =200

            # Find coordinates of all pixels below threshold
            coords = np.column_stack(np.where(gray == threshold_level))
            #print(coords)

            # Create mask of all pixels lower than threshold level
            mask = gray == threshold_level

            # Color the pixels in the mask
            image[mask] = (204, 119, 0)

            #cv2.imshow('image', image)
            cv2.waitKey()
            return coords
        output = out_fun() 
        file = open("sample.txt","w") 
        file.write(str(output)) 
        file.close() 
        #Copy and paste...will work. 






        def swap(str, i, j):
            list1 = str.split(',')
            list1[i], list1[j] = list1[j], list1[i]
            return ''.join(list1)

        f = open("sample.txt", "r")
        content = f. read()
        listx=[]
        listy=[]
        temp=0
        result = re.sub("(\d+)\s+(\d+)", r"\1,\2", content)
        result = result. splitlines()
        for i in result:
            listx.append(i)
        for x in listx:
            y=swap(x,0,1)
            y=y.replace('[','')
            y=y.replace(']','')

            y=y.replace(',','')
            listy.append(y)
        listy=str(listy)
        listy=listy.replace(",","], [")
        listy=listy.replace("'","")
        finalval = re.sub("(\d+)\s+(\d+)", r"\1,\2", listy)
        finalval=finalval.replace(' ','')
        finalval=finalval.replace('],[','], [')
        #finalval=finalval.replace('[','')
        #finalval=finalval.replace(']','')
        finalval='['+finalval+']'

        file1 = open("sample1.txt","w") 
        file1.write(str(finalval)) 
        file1.close()

        #print(finalval)




        f = open("sample1.txt", "r")
        content = f. read()


        my_list = np.array(ast.literal_eval(content))
        ind = np.lexsort((my_list[:,1], my_list[:,0]))
        sorte = my_list[ind]

        mid=int(len(sorte)/2)
        firstpart=sorte[10:mid-5]
        secondpart=sorte[mid:mid+25]

        sort1 = np.lexsort((firstpart[:,0], firstpart[:,1]))
        sorted1 = firstpart[sort1]

        sort2 = np.lexsort((secondpart[:,0], secondpart[:,1]))
        sorted2 = secondpart[sort2]

        x1=sorted1[0]
        x2=sorted2[0]
        #print('[',x1,",",x2,']')
        result1 = np.where(sorte == x1)
        result2 = np.where(sorte == x2)
        listOfCoordinates1= list(zip(result1[0], result1[1]))
        listOfCoordinates2= list(zip(result2[0], result2[1]))

        for cord1 in listOfCoordinates1[::-1]:
            cord1=cord1

        for cord2 in listOfCoordinates2[::-1]:
            cord2=cord2    
        crest1loc=cord1[0]
        crest2loc=cord2[0]



        freq=sorte[crest1loc:]
        cordstr=''.join(str(freq))
        cordstr=cordstr.replace('\n ',',')
        cordstr=cordstr.replace('  ',',')
        #print(cordstr)


        sortb=sorte[mid-20:mid+20]
        ina = np.lexsort((sortb[:,0], sortb[:,1]))
        sortc = sortb[ina]
        trough=sortc[(int(len(sortc)))-1]
        #print(trough)



        xa=x1[0]
        xb=x2[0]
        ya=x1[1]
        yb=x2[1]
        za=trough[1]

        width, height = 800, 600
        image = mpimg.imread("frame0final.png")
        line_thickness = 2
        cv2.line(image, (xa, ya), (xb, yb), (0, 255, 0), thickness=line_thickness)
        #plt.imshow(image)
        wavelength=(xb-xa)/30.76923
        #print("Wavelength=",wavelength,"cm")
        amplitude1=(za-ya)/23.913
        amplitude2=(za-yb)/23.913
        #print("Amplitude of first wave=",amplitude1,"cm","\nAmplitude of second wave=",amplitude2,"cm")

        datas=[int(wavelength)]
        print(wavelength)
        with open('dataset.csv', 'a', newline='\n') as f_object:
            writer_object = writer(f_object)
            writer_object.writerow(datas)
            f_object.close()
        cordstr=''.join(str(freq))
    except:
        continue
```
